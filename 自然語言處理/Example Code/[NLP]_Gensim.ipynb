```python
import os
from google.colab import drive
import jieba
with open ('wiki_zh.txt','r') as readfile:
  raw_data=readfile.read()
res = []
sentence = []
error_num = 0
for i, doc in enumerate(raw_data.split('<doc ')):
    #print(doc)
    articles = ''
    try:
        for j, t in enumerate(doc.strip().split('\n')[1:-1]):
            if j == 0:
                title = t
                #print(title)
            elif len(t) == 0:
                continue
            else:
                articles += t
        res.append({'id' : i, 'title' : title,'articles' : articles})
        tmp = articles.split('\n')
        for s in tmp:
          sentence.append(s)


    except Exception as e:
        error_num+=1
        continue
#print(type(sentence))
#print(sentence)
with open('wiki_sentence.txt','w') as writer:
  for i in sentence:
    terms = jieba.lcut(i, cut_all=False)
    for j in terms:
      writer.write(j+' ')
```

```python
from gensim.models import word2vec
source = 'wiki_sentence.txt'
sentences = word2vec.LineSentence(source)
model_zh = word2vec.Word2Vec(sentences, min_count=1, vector_size=100, workers=8, epochs=5, window=5, sg=1, batch_words=200)
model_zh.save("model_ZH")

```

```python
from warnings import catch_warnings
from gensim.models import word2vec
from gensim import models
from gensim.models import KeyedVectors
import pandas as pd


model_zh = models.Word2Vec.load("model_ZH")

#印出模型裡有的詞
print(model_zh.wv.index_to_key)
print(model_zh.wv.key_to_index)
print('-'*500)
num=0
#印出模型裡有的詞嵌入向量
print(type(model_zh.wv.vectors))
print(model_zh.wv.vectors)
print('-'*500)

word_vectors = pd.DataFrame(model_zh.wv.vectors, index=model_zh.wv.index_to_key)
print(word_vectors)
print('-'*500)

#儲存詞嵌入向量
word_vectors = model_zh.wv
word_vectors.save("word2vec.wordvectors")


print(model_zh.wv.get_vector('總理', norm=False))
# 前 100 相似詞
words = model_zh.wv.most_similar("美國",topn = 100)
print(words)
print('-'*500)
# 兩個詞的相關性
print(model_zh.wv.similarity("美國","韓國"))

print('-'*500)
print(model_zh.wv.most_similar(positive=['日本', '總統'], negative=['美國']))
```
