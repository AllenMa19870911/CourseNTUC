```python
import cv2
from google.colab.patches import cv2_imshow
import numpy as np
img = cv2.imread('rushmore.jpg')
cv2.line(img,(50,50),(150,150),(0,0,255),5)
cv2_imshow(img)
print('-'*70)
cv2.arrowedLine(img,(20,50),(150,150),(0,0,255),5)
cv2_imshow(img)
print('-'*70)
cv2.rectangle(img,(50,50),(150,150),(0,0,255),5)  # 繪製正方形
cv2_imshow(img)
print('-'*70)
cv2.circle(img,(150,150),100,(0,0,255),5)
cv2_imshow(img)
print('-'*70)

img2 = np.zeros((300,300,3), dtype='uint8')
cv2.ellipse(img2,(150,150),(25,200),45,0,360,(0,0,255),5)    #img, center, axes, angle, startAngle, endAngle, color, thickness
cv2.ellipse(img2,(150,150),(20,120),30,0,360,(0,255,255),5)
cv2_imshow(img2)

print('-'*70)
img = np.zeros((300,300,3), dtype='uint8')
pts = np.array([[150,50],[250,100],[150,250],[50,100]])   # 產生座標陣列
cv2.polylines(img,[pts],True,(0,0,255),5)   # 繪製多邊形
cv2_imshow(img)
```

```python
#影像放置文字和裁切
import cv2
from google.colab.patches import cv2_imshow

img = cv2.imread('rushmore.jpg')
cv2_imshow(img)
print(img.shape)
print('-'*70)
x = 100
y = 50
w = 150
h = 100

text = '('+str(x)+', '+str(y)+')'
org = (100,50)
fontFace = cv2.FONT_HERSHEY_SIMPLEX
fontScale = 0.5
color = (0,0,255)
thickness = 1
lineType = cv2.LINE_AA
cv2.putText(img, text, org, fontFace, fontScale, color, thickness, lineType)
cv2.rectangle(img,(100,50),(250,150),(0,0,255),1)
cv2_imshow(img)

print('-'*70)
crop_img = img[y:y+h+1, x:x+w+1]        # 取出陣列的範圍
cv2_imshow(crop_img)
```

```python
#影像疊加
import cv2
from google.colab.patches import cv2_imshow


img_blue = cv2.imread('rushmore.jpg')
img_green = cv2.imread('rushmore.jpg')
img_red = cv2.imread('rushmore.jpg')
img_blue[:,:,1] = 0    # 將綠色設為 0
img_blue[:,:,2] = 0    # 將紅色設為 0
img_green[:,:,0] = 0   # 將藍色設為 0
img_green[:,:,2] = 0   # 將紅色設為 0
img_red[:,:,0] = 0     # 將藍色設為 0
img_red[:,:,1] = 0     # 將綠色設為 0
cv2_imshow(img_blue)
print('-'*70)
cv2_imshow(img_green)
print('-'*70)
cv2_imshow(img_red)
print('-'*70)

output = cv2.add(img_red, img_green)  # 疊加紅色和綠色
cv2_imshow(output)
output = cv2.add(output, img_blue)
print('-'*70)
cv2_imshow(output)
print('-'*70)

img = cv2.imread('rushmore.jpg')
logo = cv2.imread('nutc.png')
logo = cv2.resize(logo, (270, 161))
print(logo.shape)
output = cv2.addWeighted(img, 0.8, logo, 0.2, 20)
cv2_imshow(output)
```

```python
#OpenCV調整亮度和對比
import cv2
from google.colab.patches import cv2_imshow
import numpy as np

img = cv2.imread('rushmore.jpg')
cv2_imshow(img)
print('-'*70)
contrast = 200 #對比
brightness = 0 #亮度
output = img * (contrast/127 + 1) - contrast + brightness # 轉換公式
output = np.clip(output, 0, 255)
output = np.uint8(output)
cv2_imshow(output)

print('-'*70)

contrast = 0 #對比
brightness = 100 #亮度
output = img * (contrast/127 + 1) - contrast + brightness # 轉換公式
output = np.clip(output, 0, 255)
output = np.uint8(output)
cv2_imshow(output)
```

```python
#OpenCV直方圖均衡化
import cv2
from google.colab.patches import cv2_imshow
import numpy as np
from matplotlib import pyplot as plt

print('灰階圖像')
img = cv2.imread('darknight.jpg')
gray_img = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
cv2_imshow(img)
print('-'*70)
print('直方圖均衡化')
equalize_img = cv2.equalizeHist(gray_img)
cv2_imshow(equalize_img)
print('-'*70)
#繪製直方圖
plt.hist(gray_img.ravel(), 256, [0, 255],label= 'original image')
plt.hist(equalize_img.ravel(), 256, [0, 255],label= 'equalize image')
plt.show()
print('-'*70)

print('Clahe均衡化')
clahe = cv2.createCLAHE() 
clahe_img = clahe.apply(gray_img)
cv2_imshow(clahe_img)
print('-'*70)
plt.hist(gray_img.ravel(), 256, [0, 255],label= 'original image')
plt.hist(clahe_img.ravel(), 256, [0, 255],label= 'clahe_img image')
plt.show()

print('-'*70)
print('彩色圖像')
original_img = cv2.imread("city.jpg")
print('調整前彩色圖像')
cv2_imshow(original_img)
print('-'*70)
img_yuv = cv2.cvtColor(original_img, cv2.COLOR_BGR2YUV)
clahe = cv2.createCLAHE() 
img_yuv[:,:,0] = clahe.apply(img_yuv[:,:,0])
#img_yuv[:,:,0] = cv2.equalizeHist(img_yuv[:,:,0])
img_bgr = cv2.cvtColor(img_yuv, cv2.COLOR_YUV2BGR)
print('調整後彩色圖像')
cv2_imshow(img_bgr)


```
