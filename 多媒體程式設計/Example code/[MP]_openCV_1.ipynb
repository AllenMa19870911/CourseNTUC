```python
import cv2
from google.colab.patches import cv2_imshow
img_blue = cv2.imread('rushmore.jpg')
img_green = cv2.imread('rushmore.jpg')
img_red = cv2.imread('rushmore.jpg')
img_blue[:,:,1] = 0    # 將綠色設為 0
img_blue[:,:,2] = 0    # 將紅色設為 0
img_green[:,:,0] = 0   # 將藍色設為 0
img_green[:,:,2] = 0   # 將紅色設為 0
img_red[:,:,0] = 0     # 將藍色設為 0
img_red[:,:,1] = 0     # 將綠色設為 0
cv2_imshow(img_blue)
print('-'*70)
cv2_imshow(img_green)
print('-'*70)
cv2_imshow(img_red)
```

```python
#改變大小
import cv2
from google.colab.patches import cv2_imshow
img = cv2.imread('rushmore.jpg')
output_1 = cv2.resize(img, (200, 200))   # 產生 200x200 的圖
output_2 = cv2.resize(img, (100, 300), cv2.INTER_LANCZOS4)   # 產生 100x300 的圖
cv2_imshow(output_1)
print('-'*70)
cv2_imshow(output_2)
```

```python
import cv2
from google.colab.patches import cv2_imshow
img = cv2.imread('rushmore.jpg')
output_0 = cv2.flip(img, 0)    # 上下翻轉
output_1 = cv2.flip(img, 1)    # 左右翻轉
output_2 = cv2.flip(img, -1)   # 上下左右翻轉
cv2_imshow(output_0)
print('-'*70)
cv2_imshow(output_1)
print('-'*70)
cv2_imshow(output_2)
```

```python
import cv2
from google.colab.patches import cv2_imshow
img = cv2.imread('rushmore.jpg')
output = cv2.transpose(img)    # 逆時針旋轉 90 度。
cv2_imshow(output)
print('-'*70)
output_ROTATE_90_CLOCKWISE = cv2.rotate(img, cv2.ROTATE_90_CLOCKWISE)
cv2_imshow(output_ROTATE_90_CLOCKWISE)
print('-'*70)
output_ROTATE_90_COUNTERCLOCKWISE = cv2.rotate(img, cv2.ROTATE_90_COUNTERCLOCKWISE)
cv2_imshow(output_ROTATE_90_COUNTERCLOCKWISE)
print('-'*70)
output_ROTATE_180 = cv2.rotate(img, cv2.ROTATE_180)
cv2_imshow(output_ROTATE_180)
```

```python
import cv2
from google.colab.patches import cv2_imshow
import numpy as np

img = cv2.imread('rushmore.jpg')
M = np.float32([[1, 0, 100], [0, 1, 100]]) # 2x3 矩陣，x 軸平移 100，y 軸平移 100
output = cv2.warpAffine(img, M, (300, 200))
cv2_imshow(output)
print('-'*70)
M = cv2.getRotationMatrix2D((150, 100), 45, 1)    # 中心點 (240, 180)，旋轉 45 度，尺寸 1
output2 = cv2.warpAffine(img, M, (300, 200))
cv2_imshow(output2)
print('-'*70)
y=img.shape[0]
x=img.shape[1]
M = cv2.getRotationMatrix2D((x/2, y/2), 60, 1)    # 以原圖中心為旋轉中心，旋轉 60 度，尺寸 1
output3 = cv2.warpAffine(img, M, (x, y))
cv2_imshow(output3)
print('-'*70)
p1 = np.float32([[0,0],[x,0],[0,y]])
p2 = np.float32([[25,25],[x,0],[0,y]])
M = cv2.getAffineTransform(p1, p2)
output4 = cv2.warpAffine(img, M, (300, 200))
cv2_imshow(output4)
```
