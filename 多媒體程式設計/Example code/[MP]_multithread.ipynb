```python
#concurrent.futures Thread
import time
from concurrent.futures import ThreadPoolExecutor 

def aa(x):
  i = 0
  while i<5:
    i = i + 1
    time.sleep(0.5)
    print('A:', i+x)

def bb():
  i = 0
  while i<50:
    i = i + 10
    time.sleep(0.5)
    print('B:', i)
aa(0)
bb()

print('-'*70)

executor = ThreadPoolExecutor()                # 建立非同步的多執行緒的啟動器
with ThreadPoolExecutor() as executor:
  executor.submit(aa, 0)
  executor.submit(bb) 

print('-'*70)

executor = ThreadPoolExecutor()                # 建立非同步的多執行緒的啟動器
with ThreadPoolExecutor() as executor:
  for i in range(5):
    executor.submit(aa,0)

print('-'*70)

executor = ThreadPoolExecutor()                # 建立非同步的多執行緒的啟動器
x=[10,20,30,40,50]
with ThreadPoolExecutor() as executor:
  executor.map(aa,x)    
```

```python
#Threading
import time
import threading

def aa(x):
  i = 0
  while i<3:
    i = i + 1
    time.sleep(0.5)
    print('A:', i+x)

def bb():
  i = 0
  while i<50:
    i = i + 10
    time.sleep(0.5)
    print('B:', i)

def cc():
  i = 0
  while i<500:
    i = i + 100
    time.sleep(0.5)
    print('C:', i)

a = threading.Thread(target=aa, args=(0,))
b = threading.Thread(target=bb)
c = threading.Thread(target=cc)

a.start()
b.start()
a.join()   # 加入等待 aa() 完成的方法
c.start()  # 當 aa() 完成後，就會開始執行 cc()
c.join()

print('-'*70)

x= [10,20,30,40,50]
threads = []
for i in range(5):
  threads.append(threading.Thread(target=aa, args=(i,)))
  threads[i].start()
```
